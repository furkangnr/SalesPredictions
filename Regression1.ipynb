####  HANDLING MISSING VALUES 

```python
tv2.isnull().sum()
```

**I decided to remove columns with more than %50 percent NaN values. Since our dataset has 100 observations, I will remove 
columns which have more than 50 NaN values. I will examine other NaN valued columns as well.**

```python
def remove_cols_with_high_NaNs(data, threshold = 0.5):
    """ This function will take a dataframe and remove its columns which have NaN values that are more than 50 
    percent of the total number of observations. Function will return a new dataframe."""
    data2 = data.copy()
    count = 0 
    for col in data2.columns:
        if ( data2[col].isnull().sum() / data2.shape[0] ) > threshold :
            count += 1
            data2.drop(col, axis = 1, inplace = True)
    print(f"{count} columns removed")
    return data2
```

```python
tv3 = remove_cols_with_high_NaNs(tv2)
```

```python
tv2.shape, tv3.shape
```

**My hypothesis is correct. This data point includes many NaN values. I will remove that data point.**

```python
row_Na = tv3[tv3["TV_Pazar"].isnull()].index
```

**Lets check whether there are any missing points exists in our data.**
